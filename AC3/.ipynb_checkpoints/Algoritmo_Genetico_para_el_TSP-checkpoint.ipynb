#Funciones de la Actividad Guiada 3

```python
#Funciones de la Actividad Guiada 3
#Se genera una solucion aleatoria con comienzo en en el nodo 0
def crear_solucion(Nodos):
  solucion = [Nodos[0]]
  for n in Nodos[1:] :
    solucion = solucion + [random.choice(list(set(Nodos) - set(solucion)))]
  return solucion

#Devuelve la distancia entre dos nodos
def distancia(a,b, problem):
  return problem.get_weight(a,b)


#Devuelve la distancia total de una trayectoria/solucion
def distancia_total(solucion, problem):
  distancia_total = 0
  for i in range(len(solucion)-1):
    distancia_total += distancia(solucion[i] ,solucion[i+1] ,  problem)
  return distancia_total + distancia(solucion[len(solucion)-1] ,solucion[0], problem)
```

#Funciones Auxiliares

```python
#Genera una poblacion inicial de soluciones de tama√±o N.
# Puede ser v√°lida la soluci√≥n aleatoria de la AG3: crear_solucion(Nodos)

# Obtener nodos
nodos = list(problem.get_nodes())

# Generar una poblaci√≥n de tama√±o N (soluciones aleatorias)
def generar_poblacion(Nodos, N):
    poblacion = []
    for _ in range(N):
        # Generar una soluci√≥n aleatoria (una permutaci√≥n de los nodos)
        solucion = random.sample(Nodos, len(Nodos))
        poblacion.append(solucion)
    return poblacion

# Tama√±o de la poblaci√≥n
N = 10  # Puedes ajustar este valor
poblacion = generar_poblacion(nodos, N)

# Ver la poblaci√≥n generada
for solucion in poblacion:
    print(solucion)
```

```python
#Evalua la poblaci√≥n y devuelve el mejor individuo
# Funci√≥n para evaluar el costo total de una soluci√≥n
def evaluar_solucion(solucion, problem):
    costo_total = 0
    for i in range(len(solucion) - 1):
        costo_total += problem.get_weight(solucion[i], solucion[i + 1])
    # A√±adir la distancia entre el √∫ltimo y el primer nodo para cerrar el ciclo
    costo_total += problem.get_weight(solucion[-1], solucion[0])
    return costo_total

# Evaluar la poblaci√≥n y devolver el mejor individuo
def encontrar_mejor_individuo(poblacion, problem):
    mejor_individuo = None
    mejor_costo = float('inf')
    
    for individuo in poblacion:
        costo = evaluar_solucion(individuo, problem)
        if costo < mejor_costo:
            mejor_costo = costo
            mejor_individuo = individuo
    
    return mejor_individuo, mejor_costo

# Tama√±o de la poblaci√≥n
N = 10
poblacion = generar_poblacion(nodos, N)

# Encontrar el mejor individuo
mejor_individuo, mejor_costo = encontrar_mejor_individuo(poblacion, problem)

# Mostrar el mejor individuo y su costo
print("Mejor Individuo:", mejor_individuo)
print("Costo del Mejor Individuo:", mejor_costo)
```

```python
#Funcion de cruce. Recibe una poblacion(lista de soluciones) y devuelve la poblaci√≥n ampliada con los hijos.
# Todos los individuos de la poblaci√≥n son selecionados para el cruce(si la poblaci√≥n es par)
# Podr√≠a aplicarse un proceso previo de selecci√≥n para elegir los individuos que se desea cruzar.


# Funci√≥n de cruce (PMX - Partially Mapped Crossover)
import random

# Funci√≥n de cruce (PMX - Partially Mapped Crossover)
def cruzar_padres(padre1, padre2):
    size = len(padre1)
    # Seleccionar dos puntos de cruce al azar
    punto1, punto2 = sorted(random.sample(range(size), 2))
    
    # Crear hijos vac√≠os
    hijo1 = [None] * size
    hijo2 = [None] * size
    
    # Copiar la parte intermedia de los padres a los hijos
    hijo1[punto1:punto2] = padre1[punto1:punto2]
    hijo2[punto1:punto2] = padre2[punto1:punto2]
    
    # Funci√≥n auxiliar para llenar el resto de los genes
    def completar_hijo(hijo, padre, padre_contrario, punto1, punto2):
        for i in range(size):
            if hijo[i] is None:
                gen = padre[i]
                while gen in hijo:
                    idx_gen = padre.index(gen)
                    gen = padre_contrario[idx_gen]
                hijo[i] = gen

    # Completar los hijos
    completar_hijo(hijo1, padre2, padre1, punto1, punto2)
    completar_hijo(hijo2, padre1, padre2, punto1, punto2)
    
    return hijo1, hijo2

# Funci√≥n de mutaci√≥n (intercambia dos nodos aleatoriamente)
def mutar(individuo):
    i, j = random.sample(range(len(individuo)), 2)
    individuo[i], individuo[j] = individuo[j], individuo[i]

# Funci√≥n para realizar el cruce sobre la poblaci√≥n
def Cruzar(poblacion, mutacion, problem):
    nueva_poblacion = []
    num_individuos = len(poblacion)
    
    # Si la poblaci√≥n es impar, se ignora el √∫ltimo individuo en el cruce
    if num_individuos % 2 != 0:
        num_individuos -= 1
    
    # Realizar el cruce para todos los individuos en pares
    for i in range(0, num_individuos, 2):
        padre1 = poblacion[i]
        padre2 = poblacion[i+1]
        
        # Cruzar los padres para generar dos hijos
        hijo1, hijo2 = cruzar_padres(padre1, padre2)
        
        # Aplicar mutaci√≥n seg√∫n la probabilidad de mutaci√≥n
        if random.random() < mutacion:
            mutar(hijo1)
        if random.random() < mutacion:
            mutar(hijo2)
        
        # A√±adir los hijos a la nueva poblaci√≥n
        nueva_poblacion.append(hijo1)
        nueva_poblacion.append(hijo2)
    
    # Devolver la poblaci√≥n original m√°s los hijos generados
    return poblacion + nueva_poblacion

# Ejemplo de uso
N = 10  # Tama√±o de la poblaci√≥n
mutacion = 0.1  # Probabilidad de mutaci√≥n (10%)
poblacion = generar_poblacion(nodos, N)

# Cruzar la poblaci√≥n
poblacion_ampliada = Cruzar(poblacion, mutacion, problem)

# Mostrar la nueva poblaci√≥n ampliada
for i, individuo in enumerate(poblacion_ampliada):
    print(f"Individuo {i + 1}: {individuo}")
```

```python
#Funcion para generar hijos a partir de 2 padres:
# Se elige el metodo de 1-punto de corte pero es posible usar otros n-puntos, uniforme, dependiendo del problema
# Funci√≥n de cruce de 1 punto de corte
def cruzar_1_punto(padre1, padre2):
    size = len(padre1)
    # Elegir un punto de corte al azar
    punto_corte = random.randint(1, size - 1)
    
    # Generar hijos utilizando el punto de corte
    hijo1 = padre1[:punto_corte] + [gen for gen in padre2 if gen not in padre1[:punto_corte]]
    hijo2 = padre2[:punto_corte] + [gen for gen in padre1 if gen not in padre2[:punto_corte]]
    
    return hijo1, hijo2

# Funci√≥n de mutaci√≥n (intercambia dos nodos aleatoriamente)
def mutar(individuo):
    i, j = random.sample(range(len(individuo)), 2)
    individuo[i], individuo[j] = individuo[j], individuo[i]

# Funci√≥n para generar descendencia a partir de dos padres
def Descendencia(padres, problem, mutacion):
    padre1, padre2 = padres
    
    # Generar dos hijos a partir de los padres utilizando cruce de 1 punto
    hijo1, hijo2 = cruzar_1_punto(padre1, padre2)
    
    # Aplicar mutaci√≥n a los hijos seg√∫n la probabilidad de mutaci√≥n
    if random.random() < mutacion:
        mutar(hijo1)
    if random.random() < mutacion:
        mutar(hijo2)
    
    return hijo1, hijo2

# Ejemplo de uso
N = 10  # Tama√±o de la poblaci√≥n
mutacion = 0.1  # Probabilidad de mutaci√≥n (10%)
poblacion = generar_poblacion(nodos, N)

# Seleccionar dos padres al azar
padre1 = poblacion[0]
padre2 = poblacion[1]

# Generar descendencia
hijo1, hijo2 = Descendencia([padre1, padre2], problem, mutacion)

# Mostrar los hijos generados
print("Padre 1:", padre1)
print("Padre 2:", padre2)
print("Hijo 1:", hijo1)
print("Hijo 2:", hijo2)
```

```python
#Para el operador de cruce 1-punto los hijos generados no son soluciones(algunos nodos se repiten y otros no est√°n)
def Factibilizar(solucion, problem):
    # Obtener todos los nodos posibles en el problema
    nodos_validos = set(problem.get_nodes())
    
    # Encontrar los nodos que est√°n repetidos en la soluci√≥n
    nodos_vistos = set()
    nodos_repetidos = []
    
    for nodo in solucion:
        if nodo in nodos_vistos:
            nodos_repetidos.append(nodo)
        else:
            nodos_vistos.add(nodo)
    
    # Encontrar los nodos que faltan en la soluci√≥n
    nodos_faltantes = list(nodos_validos - nodos_vistos)
    
    # Reemplazar los nodos repetidos por los nodos que faltan
    solucion_factible = solucion[:]
    for i in range(len(solucion_factible)):
        if solucion_factible[i] in nodos_repetidos:
            solucion_factible[i] = nodos_faltantes.pop(0)  # Reemplazar con un nodo faltante
    
    return solucion_factible

# Ejemplo de uso
padre1 = [1, 2, 3, 4, 5, 6]
padre2 = [1, 3, 5, 6, 2, 4]

# Hijo generado con repetidos
hijo1 = [1, 2, 3, 1, 5, 6]

# Factibilizar el hijo
hijo1_factible = Factibilizar(hijo1, problem)

# Mostrar el resultado
print("Hijo original:", hijo1)
print("Hijo factible:", hijo1_factible)
```

```python
#Funcion de mutaci√≥n. Se eligen dos nodos y se intercambia. Se podrian a√±adir otros operaradores
# Se hace mutaciones mutacion% de las veces
def Mutar(solucion, mutacion):
    # Probabilidad de realizar la mutaci√≥n
    if random.random() < mutacion:
        # Seleccionar dos √≠ndices aleatorios para intercambiar los nodos
        i, j = random.sample(range(len(solucion)), 2)
        
        # Intercambiar los nodos en las posiciones i y j
        solucion[i], solucion[j] = solucion[j], solucion[i]
    
    return solucion

# Ejemplo de uso
solucion = [1, 2, 3, 4, 5, 6]
mutacion = 0.3  # Probabilidad del 30% de aplicar mutaci√≥n

# Aplicar mutaci√≥n
solucion_mutada = Mutar(solucion, mutacion)

# Mostrar la soluci√≥n antes y despu√©s de la mutaci√≥n
print("Soluci√≥n original:", solucion)
print("Soluci√≥n mutada:", solucion_mutada)
```

```python
#Funcion de seleccion de la poblaci√≥n. Recibe como parametro una poblacion y
# devuelve una poblacion a la que se ha eliminado individuos poco aptos(fitness alto) y para mantener una poblacion estable de N individuos
#Se tiene en cuenta el porcentaje elitismo pasado como parametro
# Para los individuos que no son de la elite podr√≠amos usar una selecci√≥n de ruleta(proporcional a su fitness)
# Funci√≥n para calcular el fitness de un individuo (a menor costo, mejor fitness)
def calcular_fitness(individuo, problem):
    return evaluar_solucion(individuo, problem)

# Funci√≥n de selecci√≥n de la poblaci√≥n con elitismo y ruleta
def Seleccionar(problem, poblacion, N, elitismo):
    # Calcular el fitness para cada individuo en la poblaci√≥n
    poblacion_con_fitness = [(individuo, calcular_fitness(individuo, problem)) for individuo in poblacion]
    
    # Ordenar la poblaci√≥n por fitness (de menor a mayor, porque menor costo es mejor)
    poblacion_con_fitness.sort(key=lambda x: x[1])
    
    # Selecci√≥n de la elite: los mejores individuos (seg√∫n el porcentaje de elitismo)
    num_elite = int(elitismo * N)
    elite = [individuo for individuo, fitness in poblacion_con_fitness[:num_elite]]
    
    # Selecci√≥n por ruleta para el resto de los individuos
    fitness_total = sum(1 / fitness for _, fitness in poblacion_con_fitness[num_elite:])  # Invertir fitness para proporcionalidad
    seleccionados = []

    # Probabilidad proporcional al fitness
    for _ in range(N - num_elite):
        rand = random.uniform(0, fitness_total)
        acumulado = 0
        for individuo, fitness in poblacion_con_fitness[num_elite:]:
            acumulado += 1 / fitness
            if acumulado >= rand:
                seleccionados.append(individuo)
                break
    
    # Devolver la nueva poblaci√≥n con la elite y los seleccionados por ruleta
    nueva_poblacion = elite + seleccionados
    return nueva_poblacion

# Ejemplo de uso
N = 10  # Tama√±o de la poblaci√≥n a mantener
elitismo = 0.2  # Elitismo del 20% (mantener a los mejores)
poblacion_seleccionada = Seleccionar(problem, poblacion, N, elitismo)

# Mostrar la poblaci√≥n seleccionada
for i, individuo in enumerate(poblacion_seleccionada):
    print(f"Individuo {i + 1}: {individuo}")
```

#Proceso Principal

```python
#Funcion principal del algoritmo genetico
#######################################################3
import random

# Funci√≥n para generar la poblaci√≥n inicial
def generar_poblacion(Nodos, N):
    poblacion = []
    for _ in range(N):
        solucion = random.sample(Nodos, len(Nodos))  # Genera una permutaci√≥n de los nodos
        poblacion.append(solucion)
    return poblacion

# Funci√≥n para evaluar la mejor soluci√≥n en la poblaci√≥n
def evaluar_mejor_poblacion(poblacion, problem):
    mejor_solucion = None
    mejor_distancia = float('inf')
    for individuo in poblacion:
        distancia = evaluar_solucion(individuo, problem)
        if distancia < mejor_distancia:
            mejor_distancia = distancia
            mejor_solucion = individuo
    return mejor_solucion, mejor_distancia

# Funci√≥n principal del algoritmo gen√©tico
def algoritmo_genetico(problem, N=100, mutacion=0.15, elitismo=0.1, generaciones=100):
    # Obtener nodos del problema
    Nodos = list(problem.get_nodes())
    
    # Generar la poblaci√≥n inicial
    poblacion = generar_poblacion(Nodos, N)

    # Inicializamos valores para la mejor soluci√≥n
    mejor_solucion, mejor_distancia = evaluar_mejor_poblacion(poblacion, problem)

    # Condici√≥n de parada
    n = 0
    while n < generaciones:
        # Cruce de la poblaci√≥n (incluye mutaci√≥n)
        poblacion = Cruzar(poblacion, mutacion, problem)

        # Seleccionar la nueva poblaci√≥n
        poblacion = Seleccionar(problem, poblacion, N, elitismo)

        # Evaluamos la nueva poblaci√≥n
        mejor_solucion, mejor_distancia = evaluar_mejor_poblacion(poblacion, problem)

        print(f"Generaci√≥n #{n}: La mejor soluci√≥n es {mejor_solucion} con distancia {mejor_distancia}")

        # Incrementamos el n√∫mero de generaciones
        n += 1

    return mejor_solucion

# Ejemplo de uso
sol = algoritmo_genetico(problem=problem, N=500, mutacion=0.3, elitismo=0.4, generaciones=250)
print("Soluci√≥n final:", sol)
```
